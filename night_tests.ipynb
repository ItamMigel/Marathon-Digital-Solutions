```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('data/data.csv', sep=';')
```

```python
df.head(1)
```

```python
df = df.dropna()
```

```python
df = df[df['Поток 1'] != 0]
df = df[df['Поток 2'] != 0]
```

```python
df.describe()
```

```python
df.drop(columns=['Время'], inplace=True)
```

```python
df.head(3)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['Поток 1', 'Поток 2']), df['Поток 1'], test_size=0.2, random_state=42)
```

```python
import catboost

model = catboost.CatBoostRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

ans = model.predict(X_test)

```
