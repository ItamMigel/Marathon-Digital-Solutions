```python
import pandas as pd
```

```python
file_path = '../data/data.csv'
output_path = '../data/all_data.parquet'
```

```python
print(f'Processing data from {file_path} to {output_path}')
df = pd.read_csv(file_path, sep=';')
print(f'Data loaded')
```

```python
df_columns = df.columns
area_1_columns = [
    'Время',
    'Мощность МПСИ 1 кВт',
    'Мощность МШЦ 1 кВт',
    'Ток МПСИ 1 А',
    'Ток МШЦ 1 А',
    'Исходное питание МПСИ 1 т/ч',
    'Возврат руды МПСИ 1 т/ч',
    'Общее питание МПСИ 1 т/ч',
    'Расход воды МПСИ 1 PV м3/ч',
    'Расход воды МПСИ 1 SP м3/ч',
    'Расход воды МПСИ 1 CV %',
    'факт соотношение руда/вода МПСИ 1',
    'Давление на подшипник МПСИ 1 загрузка Бар',
    'Давление на подшипник МПСИ 1 разгрузка Бар',
    'Температура масла основной маслостанции подача МПСИ 1',
    'Температура масла основной маслостанции слив МПСИ 1',
    'Температура масла маслостанции электродвигатель МПСИ 1',
    'Температура масла редуктора МПСИ 1',
    'Давление на подшипник МШЦ 1 загрузка Бар',
    'Давление на подшипник МШЦ 1 разгрузка Бар',
    'Температура масла основной маслостанции подача МШЦ 1',
    'Температура масла основной маслостанции слив МШЦ 1',
    'Температура масла маслостанции электродвигатель МШЦ 1',
    'Температура масла редуктора МШЦ 1',
    'Расход извести МШЦ 1 л/ч',
    'Уровень в зумпфе 1 %',
    'Обороты насоса 1 1 %',
    'Давление в ГЦ насоса 1 1 Бар',
    'Плотность слива ГЦ 1 кг/л',
    'pH оборотной воды',
    't оборотной воды',
    'Гранулометрия 1 %',
    'Поток 1 л/мин'
]


area_2_columns = [
    'Время',
    'Мощность МПСИ 2 кВт',
    'Мощность МШЦ 2 кВт',
    'Ток МПСИ 2 А',
    'Ток МШЦ 2 А',
    'Исходное питание МПСИ 2 т/ч',
    'Возврат руды МПСИ 2 т/ч',
    'Общее питание МПСИ 2 т/ч',
    'Расход воды МПСИ 2 PV м3/ч',
    'Расход воды МПСИ 2 SP м3/ч',
    'Расход воды МПСИ 2 CV %',
    'факт соотношение руда/вода МПСИ 2',
    'Давление на подшипник МПСИ 2 загрузка Бар',
    'Давление на подшипник МПСИ 2 разгрузка Бар',
    'Обороты насоса 2 1 %',
    'Давление в ГЦ насоса 2 1 Бар',
    'Расход в ГЦ насоса 2 1 м3/ч',
    'Расход оборотной воды 2 м3/ч',
    'pH оборотной воды',
    't оборотной воды',
    'Гранулометрия 2 %',
    'Поток 2 л/мин'
]
```

```python
df_area_1 = df[area_1_columns]
df_area_2 = df[area_2_columns]
```

```python
df_area_2.columns
```

```python

df_area_1.columns = [col.replace(' 1', '').replace(' 2', '') for col in df_area_1.columns]
df_area_2.columns = [col.replace(' 1', '').replace(' 2', '') for col in df_area_2.columns]
```

```python
df_area_2.columns
```

```python
# Проверяем и выравниваем столбцы перед конкатенацией
# Находим разницу в столбцах между таблицами
missing_in_area_1 = set(df_area_2.columns) - set(df_area_1.columns)
missing_in_area_2 = set(df_area_1.columns) - set(df_area_2.columns)

# Добавляем отсутствующие столбцы и заполняем их нулями
for col in missing_in_area_1:
    df_area_1[col] = 0

for col in missing_in_area_2:
    df_area_2[col] = 0

df_all = pd.concat([df_area_1, df_area_2], axis=0)
print(f'Длина совпадает: {len(df_all.columns) == len(set(df_area_1.columns + df_area_2.columns))}')
print(f'Data concatenated')
df_all.to_parquet(output_path, index=False)
print(f'Data saved to {output_path}')
```
